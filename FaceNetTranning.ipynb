```python
from facenet_pytorch import InceptionResnetV1
import torch

# Tải mô hình FaceNet đã được huấn luyện trên dataset VGGFace2
model = InceptionResnetV1(pretrained='vggface2').eval()

# Kiểm tra xem có GPU không, nếu có thì sử dụng
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)
```

```python
import os
from PIL import Image
from torchvision import transforms
import numpy as np
import torch

transform = transforms.Compose([transforms.Resize((160, 160)), transforms.ToTensor()])

def extract_face_vector(image_path):
    image = Image.open(image_path)
    image = transform(image).unsqueeze(0).to(device)
    with torch.no_grad():
        face_embedding = model(image)
    return face_embedding.cpu().numpy()

# Đường dẫn tới thư mục chứa ảnh owner
image_dir = '/content/drive/MyDrive/IOT/face_captures'

# Lấy danh sách các tệp ảnh trong thư mục
image_paths = [os.path.join(image_dir, fname) for fname in os.listdir(image_dir) if fname.endswith(('jpg', 'png'))]

# Trích xuất vector đặc trưng cho mỗi ảnh
face_vectors = []
for img_path in image_paths:
    try:
        vector = extract_face_vector(img_path)
        face_vectors.append(vector)
    except Exception as e:
        print(f"Không thể xử lý ảnh {img_path}: {e}")

# Lưu tất cả các vector đặc trưng lại
if face_vectors:
    np.save('owner_face_vectors_esp32-cam.npy', np.vstack(face_vectors))


```
